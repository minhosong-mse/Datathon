```python
process_keywords_minimal = [
    # Wafer
    "wafer", "substrate",

    # Oxidation
    "oxidation", "sio2",

    # Photolithography
    "lithography", "photoresist",

    # Etching
    "etching", "plasma etching",

    # Deposition (Thin film)
    "deposition", "cvd", "ald",

    # Metallization
    "metallization", "interconnect",

    # Test
    "test", "reliability",

    # Packaging
    "packaging", "bump"
]
```

```python
def is_process_minimal_related(text):
    text = text.lower()
    return any(kw in text for kw in process_keywords_minimal)

semi_df["is_process_minimal"] = semi_df["full_text"].apply(is_process_minimal_related)
process_minimal_df = semi_df[semi_df["is_process_minimal"] == True].copy()

print("최소 공정 관련 논문 수:", len(process_minimal_df))
process_minimal_df.head()
```

```python
process_yearly = {}

for kw in process_keywords:
    temp = process_df[process_df["full_text"].str.contains(kw)]
    process_yearly[kw] = temp["year"].value_counts().sort_index()

process_yearly_df = pd.DataFrame(process_yearly).fillna(0).astype(int)
process_yearly_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,7))

for kw in process_yearly_df.columns:
    if process_yearly_df[kw].sum() > 0:
        plt.plot(process_yearly_df.index, process_yearly_df[kw], marker="o", label=kw)

plt.title("Semiconductor Process Keyword Trends (2021–2025)")
plt.xlabel("Year")
plt.ylabel("Paper Count")
plt.legend()
plt.grid(True)
plt.show()
```
